```python
import pandas as pd
import os 
import warnings
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
warnings.filterwarnings("ignore")
crashes_df = pd.read_csv('Output/Crashes.csv')
#type(crashes_df.DAMAGE)
```

```python
crashes_num = len(crashes_df["RD_NO"].value_counts())
```

```python
crash_type = crashes_df.groupby(['FIRST_CRASH_TYPE']).count()["RD_NO"]
crash_type = pd.DataFrame(crash_type)
plt.style.use('ggplot')
crash_type.T.plot.bar()
plt.ylabel("Number Of Crashes")
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.1))
plt.savefig("crash_type.png")
```

```python
control_device= crashes_df.groupby(['TRAFFIC_CONTROL_DEVICE']).count()["RD_NO"]
control_device = pd.DataFrame(control_device)
control_device.T.plot.bar()
plt.style.use('ggplot')
plt.ylabel("Number Of Crashes")
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.1))
plt.savefig("control_device.png")
```

```python
type_control = pd.crosstab(crashes_df.CRASH_TYPE, crashes_df.TRAFFIC_CONTROL_DEVICE)
```

```python
type_control.T.plot.bar()
plt.savefig("type_control.png")
```

```python
hit_damage = pd.crosstab(crashes_df.DAMAGE, crashes_df.HIT_AND_RUN_I)
```

```python
hit_damage.T.plot.area()
plt.savefig("hit_damage.png")
```

```python
control = pd.crosstab(crashes_df.DEVICE_CONDITION, crashes_df.TRAFFIC_CONTROL_DEVICE)

```

```python
control.T.plot.pie(x=0,y=0)
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.1))
plt.savefig("control.png")
```
